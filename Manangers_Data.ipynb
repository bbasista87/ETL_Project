```python
#Dependencies
import pandas as pd
```

```python
#Baseball Reference url for managers
url = 'https://www.baseball-reference.com/managers/'
```

```python
#View url table by panda
tables = pd.read_html(url)
tables
```

```python
#create df
df = tables[0]

df
```

```python
#drop columns
df = df.drop(columns=['Rk','G>.500','Unnamed: 21','Player/Manager'])
```

```python
#replace NaN fields with .000
df['W-L%post'].fillna(.000, inplace=True)
print(df)
```

```python
# Drop all rows with missing information
df = df.dropna(how='any')

#Drop rows with headers
df.drop(df.loc[df['Mgr']=='Mgr'].index, inplace=True)

#View data fram with dropped fields
df
```

```python
#view columns
df.columns
```

```python
#view data types
df.dtypes
```

```python
#rename columns with % and spaces
df = df.rename(columns={'Plyof App': 'PlayoffApp','W-L%':'Season_WL_PCT','W-L%post':'Post_WL_PCT'})
```

```python
#convert from object to integer
df.Yrs = df.Yrs.astype(int)
df.From = df.From.astype(int)
df.To = df.To.astype(int)
df.W = df.W.astype(int)
df.L = df.L.astype(int)
df.Ties = df.Ties.astype(int)
df.G = df.G.astype(int)
df.Wpost = df.Wpost.astype(int)
df.Lpost = df.Lpost.astype(int)
df.BestFin = df.BestFin.astype(int)
df.WrstFin = df.WrstFin.astype(int)
df.PlayoffApp = df.PlayoffApp.astype(int)
df.WSwon = df.WSwon.astype(int)
df.PennWon = df.PennWon.astype(int)
df.ASG = df.ASG.astype(int)

#convert from object to float
df.Season_WL_PCT = df.Season_WL_PCT.astype(float)
df.Post_WL_PCT = df.Post_WL_PCT.astype(float)
df.AvRk = df.AvRk.astype(float)

df.dtypes
```

```python
#filter by managers from 1990 to 2016
filter_df = df.loc[(df['From'] > 1989) & (df['From'] < 2016)] 
manager_df = filter_df.sort_values(by='From', ascending=False)
manager_df
```

```python
#split full name to first and last name columns
splitted = manager_df['Mgr'].str.split()
manager_df['name_first'] = splitted.str[0]
manager_df['name_last'] = splitted.str[1]

manager_df
```

```python
#view column array
manager_df.columns.values
```

```python
#rearrange columns and drop 'Mgr' column
columns = ['name_last', 'name_first','Yrs', 'From', 'To', 'W', 'L', 'Season_WL_PCT', 'Ties', 'G',
       'Wpost', 'Lpost', 'Post_WL_PCT', 'BestFin', 'WrstFin', 'AvRk',
       'PlayoffApp', 'WSwon', 'PennWon', 'ASG']
manager_final = manager_df.loc[:, columns]

manager_final.head()
```

```python
manager_final.to_pickle('Managers.pkl')
```
