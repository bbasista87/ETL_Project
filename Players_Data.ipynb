```python
# Dependencies
import pandas as pd
import numpy as np
```

```python
# Name of the CSV file
file = 'Resources/player.csv'
```

```python
# The correct encoding must be used to read the CSV in pandas
df = pd.read_csv(file)
print(df)
```

```python
df.columns.values
```

```python
df = df.drop(columns=[ 'death_year','death_month', 'death_day', 'death_country', 'death_state','death_city','name_given','retro_id','bbref_id'])
```

```python
debut_year= df['debut'].str[:4]
debut_year
```

```python
df['debut_year']= debut_year
```

```python
# Drop all rows with missing information
df = df.dropna(how='any')
```
