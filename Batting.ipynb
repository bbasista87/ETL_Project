```python
import pandas as pd
import os
import csv
import matplotlib
```

```python
file_batting = "Resources/Batting.csv"

df_batting = pd.read_csv(file_batting)
```

```python
df_batting.columns

del_df= df_batting.drop(columns=["G","IBB","SH","GIDP", "lgID"])
#df_batting.dtypes
```

```python
#reducing years to only 1990-2018


reduced_df = df_batting.loc[:, ["teamID", "playerID", "yearID",
                                    "AB", "R", "2B", "3B", "HR", "RBI", 
                        "SB", "CS", "BB", "SO", "HBP", "SF", "H"]]
reduced_df = del_df.loc[(
    df_batting['yearID'] >= 1990)]
reduced_df

```

```python
#calculating batting average(AVG) and On base Percentage (OBP)

reduced_df.H = reduced_df.H.astype(float)
reduced_df.AB = reduced_df.AB.astype(float)
reduced_df.BB = reduced_df.BB.astype(float)
reduced_df.HBP = reduced_df.HBP.astype(float)
reduced_df.SF = reduced_df.SF.astype(float)

#calculating OBP

OBP= (reduced_df.H + reduced_df.BB + reduced_df.HBP)/ (reduced_df.AB+ reduced_df.BB+ reduced_df.HBP+ reduced_df.SF)

#adding OBP to dataframe

reduced_df.assign(Name='OBP')

reduced_df['OBP'] = OBP

#batting average

AVG= reduced_df.H/reduced_df.AB
AVG

#adding to dataframe
reduced_df.assign(Name='AVG')

reduced_df['AVG'] = AVG

final_batting_df = reduced_df.round(3)
```
